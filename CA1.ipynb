```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("Employee_Salaries.csv")
```

```python
data
```

```python
data.info()
```

```python
data = data.drop(["Department","Division", "Overtime_Pay", "Longevity_Pay", "Grade"], axis=1)
```

```python
data
```

```python
data.isnull()
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
ohe = OneHotEncoder()
department = data["Department_Name"].values.reshape(-1, 1)
data_encoded = ohe.fit_transform(department).toarray()

encoded_df = pd.DataFrame(data_encoded, columns=ohe.get_feature_names_out(['Department_Name']))


data = pd.concat([data, encoded_df], axis=1)
```

```python
data
```

```python
data = data.drop("Department_Name", axis=1)
```

```python
data
```

```python
le = LabelEncoder()
gender_encoded = le.fit_transform(data["Gender"])
gender_encoded_df = pd.DataFrame(gender_encoded, columns=['Gender_Encoded'])
data = pd.concat([data, gender_encoded_df], axis=1)
```

```python
data
```

```python
data = data.dropna()
```

```python
data
```

```python
data = data.drop("Gender", axis=1)
```

```python
data
```

```python

X = data.iloc[:,1:20]   
Y = data.iloc[:,0]
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python

model = Sequential()

model.add(Dense(500, input_dim=19, activation='relu')) 

model.add(Dense(1, activation='sigmoid')) 
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=150, batch_size=10,  verbose=2) # 150 epoch, 10 batch size, verbose = 2
```
